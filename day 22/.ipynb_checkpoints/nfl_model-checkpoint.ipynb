```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv('first_round_qb_training_data.csv')

df = df[-df['player'].isin(['Carson Wentz', 'Trey Lance'])]
df['age'] = df['age'].astype(int)
df['seasons'] = df['seasons'].astype(int)

players = ['Kyler Murray', 'Cam Newton', 'Baker Mayfield', 'Justin Fields', 'Joe Burrow']

bad_data = df[df['player'].isin(players)]
good_data = df[-df['player'].isin(players)]

dfs = []
for player in players:
    
    frame = bad_data[bad_data['player'] == player]
    last_name = player.split(' ')[1]
    
    player_frame = pd.read_csv(f'{last_name}.csv')
    player_frame = player_frame[player_frame['Season'] == 'Career']
    frame = bad_data[bad_data['player'] == player]
    
    shared_cols = np.intersect1d(df[df['player'] == player].columns, player_frame.columns)
    
    for col in shared_cols:
        frame[col] = player_frame[col].values[0]
    
    dfs.append(frame)

df = pd.concat([good_data, pd.concat(dfs)])
```

```python
sns.countplot(data=df, x='age')
plt.title('Age Distribution for 1st Round QBs (2010-2022)')
```

```python
df.columns
```

```python
team_order = df['team'].value_counts().index
sns.countplot(data=df, x='team', order=team_order)
plt.xticks(rotation=80)
plt.title('First Round Distribution by Team')
plt.show()
```

```python
sns.histplot(data=df, x='pick')
plt.title('Pick # Distribution')
plt.show()
```

```python
for col in df.columns:
    if df[col].dtype == 'float64':
        sns.histplot(data=df, x=col)
        plt.title(f'Distribution of {col}')
        plt.show()
```

```python
continuous_df = df.select_dtypes(include=['float64'])
cont_corr = df.select_dtypes(include=['float64']).corr()
plt.figure(figsize=(10, 8))  # Adjust this size if necessary
sns.heatmap(cont_corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.1)
plt.title('Correlation Map for Continous Variables')
plt.show()
```

```python
plt.figure(figsize=(15,6))
sns.barplot(data=df.sort_values('G',ascending=False), x='player',y='G')
plt.xticks(rotation=90)
plt.title('Games Played in College by Player')
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns_plot = sns.scatterplot(data=df, x='TD%', y='Int%')

for i in range(len(df)):
    sns_plot.annotate(df['player'].apply(lambda x: x.split(' ')[1]).iloc[i], 
                          (df['TD%'].iloc[i], df['Int%'].iloc[i]), 
                          textcoords="offset points", 
                          xytext=(0,5), 
                          ha='center')
    
plt.title('TD% vs INT%')
```

```python
plt.figure(figsize=(15,8))
sns_plot = sns.scatterplot(data=df, x='G', y='TD')

for i in range(len(df)):
    sns_plot.annotate(df['player'].apply(lambda x: x.split(' ')[1]).iloc[i], 
                          (df['G'].iloc[i], df['TD'].iloc[i]), 
                          textcoords="offset points", 
                          xytext=(0,5), 
                          ha='center')

plt.title('Games Played vs Touchdowns')
```

```python
plt.figure(figsize=(16,8))
sns_plot = sns.scatterplot(data=df, x='AY/A', y='Int%')

for i in range(len(df)):
    sns_plot.annotate(df['player'].apply(lambda x: x.split(' ')[1]).iloc[i], 
                          (df['AY/A'].iloc[i], df['Int%'].iloc[i]), 
                          textcoords="offset points", 
                          xytext=(0,5), 
                          ha='center')

plt.title('AY/A vs. INT%')
```
